### Dessin un cercle

```python
# Dessine cercle
def cercle(x, pos):
    if pos > 0:
         return np.sqrt((1-x)*(1+x))
    else:
        return -np.sqrt((1-x)*(1+x))
```

```python
a = -1 # point départ
b = 1 # point arrive
Nb = 100 # nombre de points
X = np.linspace(a,b, Nb+1)
Y1 = cercle(X, 1)
Y2 = cercle(X, -1)
plt.plot(X,Y1)
plt.plot(X,Y2)
# Pour round
plt.axis('equal')
```

```python
# 3 points => Parabol => ax + bx^2 + c = f(x)
X = [2,5,7]
Y = [1,6,3]
plt.plot(X,Y, "*r")
A = np.array([[1,X[0], X[0]**2],[1,X[1], X[1]**2],[1,X[2], X[2]**2]])
b = np.array([Y[0],Y[1],Y[2]])
S = solve(A,b)
#print(S)
#print(A.dot(S))
#print(A)
def f(x): return S[0]+S[1]*x+S[2]*(x**2)

X = np.linspace(0, 10, 100)
Y = [ f(t) for t in X]
plt.plot(X, Y, "-b")
```

```python
# n points => Parabol => ax + bx^2 + c = f(x)
N = 7
X = [2,3,4,6,8,9,10]
Y = [10,2,1,1,1,2,10]
plt.plot(X,Y, "*r")

list_X = [1*[X[d]**t for t in range(N)] for d in range(N)]
list_Y = [Y[t] for t in range(N)]
A = np.array(list_X)
b = np.array(list_Y)
S = solve(A,b)

def f(x): 
    c = 0.
    for t in range(N):
        c += S[t]*(x**t)
    return c
print(f(0))
Vx = np.linspace(2, 10, 100)
Vy = [f(t) for t in Vx]
plt.plot(Vx, Vy, "-b")
plt.axis('equal')
```
