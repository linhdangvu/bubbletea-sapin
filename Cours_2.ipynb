```python
import numpy as np
import math as ma
import matplotlib.pyplot as plt
from numpy.linalg import solve # pour travailler des matrix
```

```python
### Corrige prof
X = [-5,-4,-3,-2,1,2,5,9,12,13,15]
Y = [-10,5,2,4,7,1,4,3,4,6,-2]
plt.plot(X,Y,"*r")

n = len(X)
A = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        A[i,j] = X[i]**j
b = np.array(Y)  
S = solve(A,b)

def Pn(x): return sum([S[i]*x**i for i in range(n)])

Vx = np.linspace(min(X),max(X),100)
plt.plot(Vx,Pn(Vx),"-b")
```

```python
# Integration - corig√© prof
X = [1,3]
Y = [2,6]
alpha0 = 2
alpha1 = -3

A = np.array([[1, X[0], X[0]**2, X[0]**3],
             [1, X[1], X[1]**2, X[1]**3],
             [0,1,2*X[0], 3*X[0]**2],
             [0,1,2*X[1], 3*X[1]**2]])
b = np.array([Y[0],Y[1],alpha0,alpha1])
S = solve(A,b)
def Q(x): return S[0] + S[1]*x + S[2]*x**2 + S[3]*x**3
Vx = np.linspace(0,4,100)
```

```python
plt.plot(X,Y,"*r")
plt.plot(Vx, Q(Vx), "-b")
```

```python
X = [1,3]
Y = [2,6]
alpha0 = -4
alpha1 = -6

A = np.array([[1,X[0]],[1,X[1]]])
b = np.array([Y[0],Y[1]])  
S1 = solve(A,b)
def P1(x): return S1[0]+S1[1]*x

A = np.array([[1,X[0],X[0]**2,X[0]**3],[1,X[1],X[1]**2,X[1]**3],[0,1,2*X[0],3*X[0]**2],[0,1,2*X[1],3*X[1]**2]])
b = np.array([Y[0],Y[1],alpha0,alpha1])  
S = solve(A,b)

def Q(x): return S[0] + S[1]*x + S[2]*x**2 + S[3]*x**3 
Vx = np.linspace(0,4,100)
plt.plot(X,Y,"*r")
plt.plot(Vx,P1(Vx),"-g")
plt.plot(Vx,Q(Vx),"-b")
```

```python
# PROF
X = [2.5, 5, 8]
Y = [4, 3, 4]
alpha0 = 2
alpha2 = 0

matrix = np.array([
[1, X[0], X[0] ** 2, X[0] ** 3, 0, 0, 0, 0],
[0, 1, 2 * X[0], 3 * X[0] ** 2, 0, 0, 0 ,0],
[1, X[1], X[1] ** 2, X[1] ** 3, 0, 0, 0, 0],
[0, 0, 0, 0, 1, X[1], X[1] ** 2, X[1] ** 3],
[0, 1, 2 * X[1], 3 * X[1] ** 2, 0, -1, -2* X[1], -3 * X[1] ** 2],
[0, 0, 2, 6 * X[1], 0, 0, -2, -6 * X[1]],
[0, 0, 0, 0, 0, 1, 2 * X[2], 3 * X[2] ** 2],
[0, 0, 0, 0, 1, X[2], X[2] ** 2, X[2] ** 3]
])
b = np.array([Y[0], alpha0, Y[1], Y[1], 2,4, alpha2, Y[2]])
S = solve(matrix, b)

def P(x): return S[0] + S[1] * x + S[2] * x ** 2 + S[3] * x ** 3
def Q(x): return S[4] + S[5] * x + S[6] * x ** 2 + S[7] * x ** 3

Vp = np.linspace(X[0], X[1], 100)
Vq = np.linspace(X[1], X[2], 100)
plt.plot(X, Y, "+r")
plt.plot(Vp, P(Vp), "-b")
plt.plot(Vq, Q(Vq), "-y")
plt.axis('equal')
```
